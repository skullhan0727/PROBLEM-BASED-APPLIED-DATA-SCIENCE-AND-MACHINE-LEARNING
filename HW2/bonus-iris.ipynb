### Dataset ---Iris
####  Description:
The Iris dataset was used in R.A. Fisher's classic 1936 paper, The Use of Multiple Measurements in Taxonomic Problems, and can also be found on the UCI Machine Learning Repository.

It includes three iris species with 50 samples each as well as some properties about each flower. One flower species is linearly separable from the other two, but the other two are not linearly separable from each other.

The columns in this dataset are:

Id

SepalLengthCm

SepalWidthCm

PetalLengthCm

PetalWidthCm

Species

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
sns.set_style('dark')
```

```python
df=pd.read_csv("Iris.csv")[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm","Species"]]
```

```python
df
```

## 1.Scatter matix 

```python
sns.pairplot(df, hue="Species")
```

## 2. Boxplot

```python
b1=sns.boxplot(x="Species", y="SepalLengthCm", data=df)
b1.set(title="Seciese V.S Sepal Length(cm)",ylabel="Sepal Length (cm)")
b1.set_xticklabels(["setosa","virginica","versicolor"])
plt.show()
```

```python
b2=sns.boxplot(x="Species", y="SepalWidthCm", data=df)
b2.set(title="Seciese V.S Sepal Width (cm)",ylabel="Sepal Width (cm)")
b2.set_xticklabels(["setosa","virginica","versicolor"])
plt.show()
```

## 3. violin plot

```python
v1=sns.violinplot(x="Species", y="PetalLengthCm", data=df)
v1.set(title="Seciese V.S Petal Length (cm)",ylabel="Petal Length (cm)")
v1.set_xticklabels(["setosa","virginica","versicolor"])
plt.show()
```

```python
v2=sns.violinplot(x="Species", y="PetalWidthCm", data=df)
v2.set(title="Seciese V.S Petal Width (cm)",ylabel="Petal Width (cm)")
v2.set_xticklabels(["setosa","virginica","versicolor"])
plt.show()
```

## 4.3D scatter plot

```python
from mpl_toolkits.mplot3d import Axes3D
plt.figure()

setosa= np.array(df[df["Species"]=="Iris-setosa"][["SepalLengthCm","PetalLengthCm","PetalWidthCm"]])
virginica= np.array(df[df["Species"]=="Iris-virginica"][["SepalLengthCm","PetalLengthCm","PetalWidthCm"]])
versicolor=np.array(df[df["Species"]=="Iris-versicolor"][["SepalLengthCm","PetalLengthCm","PetalWidthCm"]])

data=(setosa,virginica,versicolor)
colors = ("red", "green", "blue")
groups = ("setosa", "virginica", "versicolor") 


fig = plt.figure()
ax = fig.gca(projection='3d')

for data, color, group in zip(data,colors, groups):
    x=data[:,0]
    y=data[:,1]
    z=data[:,2]
    ax.scatter(x, y, z, alpha=0.8, c=color, edgecolors='none', s=30, label=group)

plt.title('3D scatter plot')
ax.set_xlabel("Sepal Length (cm)")
ax.set_ylabel("Petal Length (cm)")
ax.set_zlabel("Sepal Length (cm)")
plt.legend(loc=2)
plt.show()
```
